```
import caffe
import numpy as np
import cv2, pylab
```

```
model = 'deploy.prototxt'
weights = 'fcn8s-heavy-pascal.caffemodel'

caffe.set_mode_gpu()
caffe.set_device(0)

net = caffe.Net(model, weights, caffe.TEST)
```

```
img = cv2.imread('/content/2007_000733.jpg')
R = img[..., 2]
G = img[..., 1]
B = img[..., 0]
im_input = np.array([[R,G,B]])

net.blobs['data'].reshape(*im_input.shape)
net.blobs['data'].data[...] = im_input

im_output = net.forward()['score'][0]
```

```
pylab.imshow(im_output.argmax(axis=0))
```
